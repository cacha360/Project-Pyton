Для расчета общего дохода по каждому продукту, суммируйте произведение количества и цены за единицу для каждой строки данных.

```python
import csv
import random
from datetime import datetime, timedelta

# Функция для генерации случайных данных, насколько я понял ее нужно написать.
def generate_fake_data(filename, num_records):
    products = ['Товар A', 'Товар B', 'Товар C', 'Товар D']
    start_date = datetime.now() - timedelta(days=30)  # Последние 30 дней
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['ID продажи', 'Дата продажи', 'Продукт', 'Количество', 'Цена за единицу'])
        for i in range(1, num_records + 1):
            product = random.choice(products)
            quantity = random.randint(1, 10)
            price = round(random.uniform(10, 100), 2)  # Случайная цена от 10 до 100
            sale_date = start_date + timedelta(days=random.randint(0, 30))
            writer.writerow([i, sale_date.strftime('%Y-%m-%d'), product, quantity, price])

# Генерирую 100 записей
generate_fake_data('data.csv', 100)
```

```python
import csv

# Функция для чтения и вывода содержимого файла CSV
def read_csv_file(filename):
    with open(filename, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            print(row)

# Читаю и вывожу содержимое файла data.csv, чтобы удостовериться, что сгенерировалось верно
read_csv_file('data.csv')
```

```python
import csv
from collections import defaultdict

# Функция для расчета общего дохода по продуктам
def calculate_total_revenue(filename):
    revenue = defaultdict(float)  # Словарь для хранения дохода по продуктам

    with open(filename, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            product = row['Продукт']
            quantity = int(row['Количество'])
            price = float(row['Цена за единицу'])
            total_income = quantity * price
            revenue[product] += total_income

    return revenue

# Вывожу общий доход
total_revenue = calculate_total_revenue('data.csv')
for product, income in total_revenue.items():
    print(f'Общий доход по продукту "{product}": {income:.2f} руб.')
```

```python
import json

# Функция для конвертации данных из CSV в JSON
def convert_csv_to_json(csv_filename, json_filename):
    data = []
    with open(csv_filename, mode='r', newline='', encoding='utf-8') as csv_file:
        reader = csv.DictReader(csv_file)
        for row in reader:
            data.append(row)

    with open(json_filename, mode='w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)

# Конвертируем данные из CSV в JSON
convert_csv_to_json('data.csv', 'data.json')
print("Данные успешно конвертированы в формат JSON и сохранены в файл data.json.")
```

```python
import json

# Функция для чтения и вывода содержимого файла JSON
def read_json_file(filename):
    with open(filename, mode='r', encoding='utf-8') as file:
        data = json.load(file)
        for entry in data:
            print(entry)

# Проверяю что же там сконвертировалось
read_json_file('data.json')
```

# Задание 3: Сетевые запросы

Используйте библиотеку `requests` для написания скрипта, который делает HTTP-запрос к API https://jsonplaceholder.typicode.com/posts и извлекает список постов.

Импортируйте библиотеку requests и создайте объект requests.Session().


Используйте метод get() объекта Session для отправки HTTP-запроса к указанному API и извлеките список постов, используя атрибут .json() для десериализацииответа в Python-объект.

```python
import requests

# Заметки делаю для себя, чтобы перед ассесментом вспомнить, что я делал

# Создание объекта сессии
session = requests.Session()

# Отправка GET-запроса к API
response = session.get("https://jsonplaceholder.typicode.com/posts")

# Десериализация ответа в Python-объект
posts = response.json()

# Вывод списка постов
for post in posts:
    print(f"ID: {post['id']}, Title: {post['title']}")
```

# Задание 4: Сетевые запросы

Проанализируйте полученные данные и найдите пост с максимальным количеством комментариев, используя соответствующий запрос к API для комментариев (https://jsonplaceholder.typicode.com/comments).

Используйте библиотеку requestsдля отправки HTTP-запроса к API комментариев, чтобы получить список комментариев для каждого поста.

Проанализируйте каждый пост, подсчитав количество комментариев к нему, и найдите пост с максимальным количеством комментариев.

```python
import requests

# Заметки для себя, чтобы вспомнить перед ассесментом, что я делал

# Создание объекта сессии
session = requests.Session()

# Получение списка постов
posts_response = session.get("https://jsonplaceholder.typicode.com/posts")
posts = posts_response.json()

# Получение списка комментариев
comments_response = session.get("https://jsonplaceholder.typicode.com/comments")
comments = comments_response.json()

# Создание словаря для хранения количества комментариев для каждого поста
comments_count = {post['id']: 0 for post in posts}

# Подсчет количества комментариев для каждого поста
for comment in comments:
    post_id = comment['postId']
    comments_count[post_id] += 1

# Поиск поста с максимальным количеством комментариев
max_comments_post_id = max(comments_count, key=comments_count.get)
max_comments_count = comments_count[max_comments_post_id]
max_comments_post = next(post for post in posts if post['id'] == max_comments_post_id)

print(f"Пост с максимальным количеством комментариев:")
print(f"ID: {max_comments_post['id']}, Title: {max_comments_post['title']}, Comments: {max_comments_count}")
```
