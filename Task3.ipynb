# Задание 1: Словари и множества

Написать программу, которая подсчитывает количество вхождений каждого символа в заданной строке и выводит результат в виде словаря.

```python
def count_chars(string):
    char_count = {}
    for char in string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count


string = input("Введите строку: ")
print(count_chars(string))
```

# Задание 2: Словари и множества

Написать программу, которая создает множество из списка чисел, введенных пользователем, и выводит его сначала в неизменном виде, а затем удаляет дубликаты и выводит результат без дублей.

На самом деле в множестве не может быть повторяющихся чисел. А вот в списке может.

```python
def create_set(numbers):
    num_set = set(numbers)

    print("Множество в неизменном виде:", numbers)
    print("Множество без дублей:", num_set)


numbers = input("Введите список чисел через пробел: ")
numbers = [int(x) for x in numbers.split()]

create_set(numbers)
```

Узнаю расположение текущей директории, чтобы знать куда загружать исходный файл изображения и куда будет сохранен новый файл.

```python
import matplotlib.pyplot as plt
from PIL import Image

# Нужно указывать только имя файла и расширение иначе программа не сможет корректно сохранить новое изображение.
filename = input("Введите имя файла изображения: ")

try:
    image = Image.open(filename) # При помощи open открываем изображение для работы с ним.
except FileNotFoundError:
    print("Файл не найден. Пожалуйста, проверьте имя файла.")
    exit()
# Программа работает при нахождении файла в определенной ранее директории.
# При помощи конструкции выше, работает при нахождении файла в директории ниже, хоть и выдает ошибку.

plt.imshow(image)
plt.show() # Проверяю, что программа видит файл. Показывает значит видит.

original_width, original_height = image.size

new_width = original_width // 2
new_height = original_height // 2
resized_image = image.resize((new_width, new_height))

plt.imshow(resized_image)
plt.show()
# Проверяю, что программа видит файл. Показывает значит видит.
# Хоть изображение выглядит идентично, при проверки по свойствам изменения прошли успешно.

new_filename = f"resized_{filename}"
resized_image.save(new_filename)

new_width, new_height = resized_image.size

print(f"Размер исходного изображения: {original_width}x{original_height}")
print(f"Размер измененного изображения: {new_width}x{new_height}")
```
